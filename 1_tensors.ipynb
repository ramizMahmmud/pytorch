```python
# Basic Multiplication
x = torch.tensor([1,2,3,4])

print(x+100)
print(x-20)
print(x*10)
print(x/2)
```

```python
# Multiplication (Dot product)
x = torch.tensor([1, 2, 3])
y = torch.tensor([5, 6 ,7])
print(x.shape)
print(y.shape)
```

```python
print(x*y)
```

```python
x = torch.tensor([[1,2,3], [5,6,7]])
y = torch.tensor([[2,3, 5], [2, 3, 2]])

print(f"Order of x: {x.shape}.\nOrder of y: {y.shape}")
```

```python
# for matrix multiplication order matrices must be (m x n) and (n x p)
# order of result of multiplication (m x p)
z = torch.matmul(x, y.T) # .T is used for transpose
print(z, z.shape)
```

Finding min, max, mean

```python
x = torch.arange(start=0, end= 100, step= 3)
print(x)
```

```python
print(f"Minimum: {x.min()}")
print(f"Max: {x.max()}")
print(f"Sum: {x.sum()}")
```

```python
# Mean/ average works only on float data type

Mean = x.type(torch.float32).mean()
print(Mean)
```
