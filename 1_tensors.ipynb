```python
# Basic Multiplication
x = torch.tensor([1,2,3,4])

print(x+100)
print(x-20)
print(x*10)
print(x/2)
```

```python
print(x*y)
```

```python
x = torch.tensor([[1,2,3], [5,6,7]])
y = torch.tensor([[2,3, 5], [2, 3, 2]])

print(f"Order of x: {x.shape}.\nOrder of y: {y.shape}")
```

```python
# for matrix multiplication order matrices must be (m x n) and (n x p)
# order of result of multiplication (m x p)
z = torch.matmul(x, y.T) # .T is used for transpose
print(z, z.shape)
```

Finding min, max, mean

```python
x = torch.arange(start=0, end= 100, step= 3)
print(x)
```

```python
print(f"Minimum: {x.min()}")
print(f"Max: {x.max()}")
print(f"Sum: {x.sum()}")
```

```python
# Mean/ average works only on float data type

Mean = x.type(torch.float32).mean()
print(Mean)
```

```python
# position of max and min data
print(f"Index of max value: {x.argmax()}")
print(f"Index of min value: {x.argmin()}")
```

```python
print(x.dtype)
```

```python
x = x.type(torch.int32)
print(x.dtype)
```

```python
# Reshaping, stacking, squeezing, unsqueezing
x = torch.arange(1,10,1)
x_reshaped = x.reshape(3,3)
print(x_reshaped)
```

```python
x_view = x.view(3,3)
x_view
```

```python
print(x.ndim)
print(x.shape)
```

```python
# after changing view changes main tensor, reshape does not
x_view[0][0] =21
x_reshaped[1][1] = 99999
print(x)
print(x_reshaped)
print(x_view)
```

```python
x = torch.tensor([[1, 2, 3]])
x_squeezed = x.squeeze()
print(x_squeezed)
print(x_squeezed.shape)
```

```python
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(x_unsqueezed)
print(x_unsqueezed.shape)
```
