```python
MATRIX = torch.tensor([[1,2,3],
                       [4,5,6],
                       [12, 15, 18],
                       [20, 22 ,25]])
print(MATRIX)
print(MATRIX.ndim)
print(MATRIX.shape)
```

```python
print(MATRIX[0])
```

```python
# print 15 from MATRIX
# row = 2 (satrt from 0)
# column = 1 (strat from 0)

print(MATRIX[2][1])
```

```python
TENSOR = torch.tensor([[[1,2,3],
                        [4,5,6],
                        [10,12,14],
                         [15,16,19]]
                       ])
```
